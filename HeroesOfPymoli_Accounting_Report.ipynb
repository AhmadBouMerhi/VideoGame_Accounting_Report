```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchasedata_df = pd.read_csv(file_to_load)
purchasedata_df.head()
```

## Player Count

* Display the total number of players


```python
# Calculate the Number of Unique Players
player_demographics = purchasedata_df.loc[:,["Gender", "SN", "Age"]] 
# Drop duplicate players
player_demographics = player_demographics.drop_duplicates()
number_players = player_demographics.count()[0]


# Display the total number of players
pd.DataFrame({"Total Players": [number_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# The mean of the Price
average = purchasedata_df["Price"].mean()
#Total Revenue
total_revenue = purchasedata_df["Price"].sum()
#Total Number of Purchases
purchase_count = purchasedata_df["Purchase ID"].count()
# Calculate the number of unique Item Name in the DataFrame
unique_items = len(purchasedata_df["Item Name"].unique())
```

```python
# Place all of the data of unique items, average price, etc.. into a summary DataFrame 
summary_table_df = pd.DataFrame({"Number of Unique Items" : unique_items,
                                  "Average Price" : [average],
                                  "Number of Purchases" : [purchase_count],
                                  "Total Revenue" : [total_revenue]})

# Round the numbers to 2 decimal places
summary_table_df = summary_table_df.round(2)
# Clean up columns with numbers
summary_table_df["Average Price"] = summary_table_df["Average Price"].map("${:,.2f}".format)
summary_table_df["Total Revenue"] = summary_table_df["Total Revenue"].map("${:,.2f}".format)

summary_table_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Put the existing players in bins
player_demographics["Age Ranges"] = pd.cut(player_demographics["Age"], age_bins, labels=group_names)

# Calculate the Numbers Age group
age_demographics_totals = player_demographics["Age Ranges"].value_counts()
# Calculate the Percentages by Age group
age_demographics_percents = age_demographics_totals / number_players * 100
# Put data in data frame
age_demographics = pd.DataFrame({"Total Count": age_demographics_totals, 
                                 "Percentage of Players": age_demographics_percents})

# Round to two decimal places
age_demographics = age_demographics.round(2)

age_demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#calculate the top spenders

top_spenders = purchasedata_df[['SN','Price']]

top_spenders_df = top_spenders.groupby('SN').count().sort_values(by = ['Price'], ascending = False)

top_spend_total = purchasedata_df.groupby('SN')['Price'].sum()

top_spend_count = top_spenders_df['Price']

top_spend_avg = round(top_spend_total/top_spend_count,2)

top_spenders_df['Average Purchase Price'] = top_spend_avg
top_spenders_df['Total Purchase Value'] = top_spend_total

top_spenders_df = top_spenders_df.rename(columns = {'Price': 'Purchase Count'})

top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Calculate the most popular items
pop_items = purchasedata_df[["Item ID","Item Name","Price"]]
popular_items = pop_items.groupby(["Item ID", "Item Name"])
popular_items = popular_items.count().sort_values(by = "Price", ascending = False)

total_purchase_value = pop_items.groupby(["Item ID", "Item Name"])["Price"].sum()
purchase_count = pop_items.groupby("Item Name").count()

popular_items["Item Price"] = total_purchase_value/popular_items['Price']
popular_items["Total Purchase Value"] = total_purchase_value
popular_items = popular_items.rename(columns = {"Price":"Count"})

popular_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Calculate the most profitable Items
most_profitable = popular_items.sort_values(by ='Total Purchase Value', ascending = False)
most_profitable.head()
```
